```python
import matplotlib.pyplot as plt
import numpy as np
```

# Ploting small scale graph

```python
xs = [0.3, 1, 2.2, 3.5]
ys = [1.9, 3.3, 0.3, 3.2]
fig,ax = plt.subplots(1,1)
ax.plot(np.array([0.3,1]),np.array([1.9,3.3]),'k--')
ax.plot([1,2.2],[3.3,0.3],'k--')
ax.plot([1,3.5],[3.3,3.2],'k--')
ax.plot([0.3,2.2],[1.9,0.3],'k--')
#ax.plot([1, 2, 3, 1.5], [1, 4, 9, 8], 'yo',label='stay points')
ax.plot(xs,ys,'go',label='stay points')
ax.set_xlabel("x(m)")
ax.set_ylabel("y(m)")
ax.legend()
```

## Ploting greedy solution to small scale

```python
xs = [0.3, 1, 2.2, 3.5]
ys = [1.9, 3.3, 0.3, 3.2]
charger_x = [0.5,0.5,1.5,3.5,1.5,2.5]
charger_y = [2.5,1.5,3.5,3.5,2.5,0.5]
fig,ax = plt.subplots(1,1)
ax.plot(np.array([0.3,1]),np.array([1.9,3.3]),'k--')
ax.plot([1,2.2],[3.3,0.3],'k--')
ax.plot([1,3.5],[3.3,3.2],'k--')
ax.plot([0.3,2.2],[1.9,0.3],'k--')
#ax.plot([1, 2, 3, 1.5], [1, 4, 9, 8], 'yo',label='stay points')
ax.plot(xs,ys,'go',label='stay points')
ax.plot(charger_x,charger_y,'rx',label="chargers")
ax.legend()
```

# Ploting bar plot on variated $P_s$ (Small scale)

```python
bar_width=0.35
names_1 = np.arange(2,6)
greedy_chargers = [6,5,4,4]
pso_chargers = [4,4,3,3]
fig,ax = plt.subplots(1,1)
ax.bar(names_1,greedy_chargers,bar_width,label='Greedy Heuristic')
ax.bar(names_1 + bar_width,pso_chargers,bar_width,label='Pso')
ax.set_xlabel("$P_s$ (w)")
ax.set_ylabel("Number of Chargers placed")
ax.set_xticks(names_1 + bar_width/2)
ax.set_xticklabels((2, 3, 4, 5))
ax.legend()
```

# Ploting large scale graph information from file

```python
with open(graph_info_path,'r') as f:
    lines=f.readlines()
    x_str = lines[1].split(',')
    x_str = x_str[0:len(x_str)-1]
    x_coordinate = []
    for item in x_str:
        x_coordinate.append(float(item))
    y_str = lines[2].split(',')
    y_str = y_str[0:len(y_str)-1]
    y_coordinate = []
    for item in y_str:
        y_coordinate.append(float(item))
    print(x_coordinate)
    print(y_coordinate)
    fig,ax = plt.subplots(1,1)
    ax.plot(x_coordinate,y_coordinate,'go',label='staypoints')
    lines = lines[4:len(lines)]
    for line in lines:
        line_split = line.split(',')
        x_1 = float(line_split[0])
        y_1 = float(line_split[1])
        x_2 = float(line_split[2])
        y_2 = float(line_split[3])
        ax.plot([x_1,x_2],[y_1,y_2],'k--')
    ax.legend()
f.close()
```
